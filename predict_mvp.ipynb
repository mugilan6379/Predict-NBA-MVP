```python
import pandas as pd
```

```python
mvp = pd.read_csv('player_mvp_stats.csv')
mvp.head()
```

```python
del mvp['Unnamed: 0']
```

```python
mvp.isnull().sum()
```

```python
mvp[mvp['3P%'].isnull()][['Player','3PA']]
```

```python
mvp=mvp.fillna(0)
```

```python
mvp.columns
```

```python
predictors = ['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year', 'W', 'L', 'W/L%', 'GB', 'PS/G',
       'PA/G', 'SRS']
```

```python
train = mvp[(mvp['Year'])<2021]
test = mvp[mvp['Year']==2021]
```

```python
train[predictors]
```

```python
from sklearn.linear_model import Ridge

reg=Ridge(alpha=.1)
```

```python
reg.fit(train[predictors],train['Share'])
```

```python
predictions =reg.predict(test[predictors])
predictions=pd.DataFrame(predictions,columns=['Prediction of Share'],index=test.index)
predictions
```

```python
combo = pd.concat([test[['Player','Share']],predictions],axis=1)
combo
```

```python
combo.sort_values('Share',ascending=False).head()
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(combo['Share'],combo['Prediction of Share'])
```
