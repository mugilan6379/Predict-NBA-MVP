```python
def find_ap(combo):
    actual = combo.sort_values('Share',ascending=False).head()
    predicted = combo.sort_values('Prediction of Share',ascending=False)
    ps=[]
    matched=0
    checked=1
    for index,row in predicted.iterrows():
        if row['Player'] in actual['Player'].values:
            matched+=1
            ps.append(matched/checked)
        checked+=1
    return sum(ps)/len(ps)
```

```python
find_ap(combo)
```

```python
aps=[]
all_predictions=[]
years=list(range(1991,2022))
for year in years[5:]:
    train = mvp[mvp['Year']<year]
    test = mvp[mvp['Year']==year]
    reg.fit(train[predictors],train['Share'])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions,columns=['Prediction of Share'],index=test.index)
    combo = pd.concat([test[['Player','Share']],predictions],axis=1)
    combo=combo.sort_values('Share',ascending=False)
    print(year,combo)
    all_predictions.append(combo)
    aps.append(find_ap(combo))
```

```python
def add_ranks(combo):
    combo = combo.sort_values('Share',ascending=False)
    combo['Rank'] = list(range(1,combo.shape[0]+1))
    combo = combo.sort_values('Prediction of Share',ascending=False)
    combo['Predicted_Rank'] = list(range(1,combo.shape[0]+1))
    combo['Rank_Diff'] = combo['Rank'] - combo['Predicted_Rank']
    return combo
```

```python
ranking = add_ranks(all_predictions[1])
ranking[(ranking['Rank'] < 6)].sort_values('Rank_Diff',ascending=False)
```

```python
def backtest(mvp,model,year,predictors):
    aps=[]
    all_predictions=[]
    for year in years[5:]:
        train = mvp[mvp['Year']<year]
        test = mvp[mvp['Year']==year]
        model.fit(train[predictors],train['Share'])
        predictions = reg.predict(test[predictors])
        predictions = pd.DataFrame(predictions,columns=['Prediction of Share'],index=test.index)
        combo = pd.concat([test[['Player','Share']],predictions],axis=1)
        combo=add_ranks(combo)
        all_predictions.append(combo)
        aps.append(find_ap(combo))
    return sum(aps)/len(aps),aps,pd.concat(all_predictions)
```

```python
mean_aps, aps, all_predictions = backtest(mvp,reg,years[5:],predictors)
mean_aps
```
